```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df_tickets = pd.read_csv('/content/customer_support_tickets.csv')
display(df_tickets.head())
```

```python
print(f"Missing values in 'Ticket Subject': {df_tickets['Ticket Subject'].isnull().sum()}")
print(f"Missing values in 'Ticket Description': {df_tickets['Ticket Description'].isnull().sum()}")
```

```python
df_tickets['Ticket Subject'] = df_tickets['Ticket Subject'].fillna('')
df_tickets['Ticket Description'] = df_tickets['Ticket Description'].fillna('')
display(df_tickets.head())
```

```python
import string

def clean_text(text):
    """
    Performs several text cleaning operations on the input string.
    - Converts text to lowercase.
    - Removes punctuation.
    - Strips extra whitespace.
    """
    text = str(text).lower()
    text = ''.join([char for char in text if char not in string.punctuation])
    text = text.strip()
    return text

print("The `clean_text` function has been defined.")
```

```python
df_tickets['Ticket Subject'] = df_tickets['Ticket Subject'].apply(clean_text)
df_tickets['Ticket Description'] = df_tickets['Ticket Description'].apply(clean_text)
display(df_tickets.head())
```

```python
display(df_tickets[['Ticket Subject', 'Ticket Description']].head())
```

```python
df_tickets['combined_text'] = df_tickets['Ticket Subject'] + ' ' + df_tickets['Ticket Description']
display(df_tickets[['Ticket Subject', 'Ticket Description', 'combined_text']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Instantiate TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the 'combined_text' column
tfidf_matrix = tfidf_vectorizer.fit_transform(df_tickets['combined_text'])

print("TF-IDF vectorization complete.")
```

```python
feature_names = tfidf_vectorizer.get_feature_names_out()
print("Number of features:", len(feature_names))
print("Sample feature names:", feature_names[:10])

# Display a sample of the TF-IDF matrix
print("\nTF-IDF matrix shape:", tfidf_matrix.shape)
print("Sample TF-IDF vector for the first document:\n", tfidf_matrix[0].toarray()[:, :10])
```

## Train and Evaluate Ticket Type Classifier

### Subtask:
Train a classification model (e.g., Logistic Regression) using the training data for 'Ticket Type'. After training, make predictions on the test set and evaluate the model's performance using metrics like accuracy and a classification report.


```python
import nltk
nltk.download('stopwords', quiet=True)
print("NLTK stopwords downloaded.")
```

```python
import string
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def clean_text(text):
    """
    Performs several text cleaning operations on the input string.
    - Converts text to lowercase.
    - Removes punctuation.
    - Removes stopwords.
    - Strips extra whitespace.
    """
    text = str(text).lower()
    text = ''.join([char for char in text if char not in string.punctuation])

    # Remove stopwords
    words = text.split()
    text = ' '.join([word for word in words if word not in stop_words])

    text = text.strip()
    return text

print("The `clean_text` function has been updated to include stopword removal.")
```

## Re-applying cleaning, feature extraction, and model training due to updated text cleaning

```python
df_tickets['Ticket Subject'] = df_tickets['Ticket Subject'].apply(clean_text)
df_tickets['Ticket Description'] = df_tickets['Ticket Description'].apply(clean_text)
display(df_tickets.head())
```

```python
df_tickets['combined_text'] = df_tickets['Ticket Subject'] + ' ' + df_tickets['Ticket Description']
display(df_tickets[['Ticket Subject', 'Ticket Description', 'combined_text']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Instantiate TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the 'combined_text' column
tfidf_matrix = tfidf_vectorizer.fit_transform(df_tickets['combined_text'])

print("TF-IDF vectorization complete with updated text.")
```

```python
feature_names = tfidf_vectorizer.get_feature_names_out()
print("Number of features:", len(feature_names))
print("Sample feature names:", feature_names[:10])

# Display a sample of the TF-IDF matrix
print("\nTF-IDF matrix shape:", tfidf_matrix.shape)
print("Sample TF-IDF vector for the first document:\n", tfidf_matrix[0].toarray()[:, :10])
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_ticket_type = label_encoder.fit_transform(df_tickets['Ticket Type'])

print("Target variable 'Ticket Type' has been numerically encoded.")
print("Encoded labels (first 5):", y_ticket_type[:5])
```

```python
from sklearn.model_selection import train_test_split

X_train_type, X_test_type, y_train_type, y_test_type = train_test_split(tfidf_matrix, y_ticket_type, test_size=0.2, random_state=42)

print(f"X_train_type shape: {X_train_type.shape}")
print(f"X_test_type shape: {X_test_type.shape}")
print(f"y_train_type shape: {y_train_type.shape}")
print(f"y_test_type shape: {y_test_type.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

logistic_model_type = LogisticRegression(max_iter=1000, random_state=42)
logistic_model_type.fit(X_train_type, y_train_type)

y_pred_type = logistic_model_type.predict(X_test_type)

accuracy_type = accuracy_score(y_test_type, y_pred_type)
report_type = classification_report(y_test_type, y_pred_type)

print(f"Logistic Regression Model Accuracy (Ticket Type): {accuracy_type:.4f}")
print("\nClassification Report (Ticket Type):\n", report_type)
```
